```python
df = pd.read_csv('/content/drive/MyDrive/DPhi Datathon/Train_Dataset.csv')
df.head()
```

```python
df = df[df['headline'] != '']
X = df.drop('is_sarcastic' , axis = 1)
y = df['is_sarcastic']
```

```python
from sklearn.model_selection import train_test_split , cross_val_score
import numpy as np
np.random.seed(42)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape , X_valid.shape , y_train.shape , y_valid.shape
```

```python
from sklearn.model_selection import train_test_split , cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB

# import hyper parameter tuning
from sklearn.model_selection import RandomizedSearchCV , GridSearchCV

# import evaluation metrics
from sklearn.metrics import confusion_matrix , classification_report , precision_score , f1_score , recall_score
from sklearn.metrics import plot_roc_curve ,accuracy_score
```

```python
x_train_obj = X_train['headline'].apply(lambda row: TextBlob(row).sentiment)
X_train['Polarity'] = [obj.polarity for obj in x_train_obj.values]
X_train['Subjectivity'] = [obj.subjectivity for obj in x_train_obj.values]

x_valid_obj = X_valid['headline'].apply(lambda row: TextBlob(row).sentiment)
X_valid['Polarity'] = [obj.polarity for obj in x_valid_obj.values]
X_valid['Subjectivity'] = [obj.subjectivity for obj in x_valid_obj.values]
```

```python
stop_words = nltk.corpus.stopwords.words('english')
stop_words.remove('no')
stop_words.remove('not')
stop_words.remove('but')
```

```python
ps = nltk.porter.PorterStemmer()

def simple_text_preprocessor(document): 
    document = str(document).lower()

    document = re.sub(r'[^a-zA-Z]',r' ', document)
    document = re.sub(r'nbsp', r'', document)
    document = re.sub(' +', ' ', document)
    
    document = ' '.join([ps.stem(word) for word in document.split()])
    
    document = ' '.join([word for word in document.split() if word not in stop_words])
    
    return document

stp = np.vectorize(simple_text_preprocessor)
```

```python
X_train['Clean Review'] = stp(X_train['headline'].values)
X_valid['Clean Review'] = stp(X_valid['headline'].values)

X_train.head()
```

```python
X_train_clean = X_train.drop(['headline' , 'Clean Review'], axis=1).reset_index(drop=True)
X_valid_clean = X_valid.drop(['headline' , 'Clean Review'], axis=1).reset_index(drop=True)

X_train_clean.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(min_df=0.0, max_df=1.0, ngram_range=(1, 1) , max_features=6000)
X_traincv = cv.fit_transform(X_train['Clean Review']).toarray()
X_traincv = pd.DataFrame(X_traincv, columns=cv.get_feature_names())

X_validcv = cv.transform(X_valid['Clean Review']).toarray()
X_validcv = pd.DataFrame(X_validcv, columns=cv.get_feature_names())
X_traincv.head()
```

```python
X_train_comb = pd.concat([X_train_clean, X_traincv], axis=1)

X_train_comb.head()
```

```python
X_valid_comb = pd.concat([X_valid_clean, X_validcv], axis=1)
```

```python
models = {'Logistic Regression' : LogisticRegression() , 
         'KNN' : KNeighborsClassifier() ,
         'Random Forests' : RandomForestClassifier() ,
         'Desicion Tree' : DecisionTreeClassifier()}

# Fitting and checkin score of models
def prep_models(models , X_train , X_test , y_train  , y_test):
    """
        fitting and testing different models 
        models = dict of the required models
    """
    
    # set the random seed
    np.random.seed(45)
    
    #scores dict 
    mod_scores = {}
    
    for name , model in models.items():
        model.fit(X_train , y_train) # fitting the model
        pred = model.predict(X_test)
        # evaluating and adding the scores to the dict
        mod_scores[name] = {'Precision' : precision_score(pred , y_test ) , 
                           'Recall' : recall_score(pred , y_test ) , 
                           'F1 Score' : f1_score(pred , y_test )  , 
                           'Accuracy' : accuracy_score(pred , y_test )}
    return mod_scores
```

```python
model_scores = prep_models(models , 
                           X_train = X_train_comb ,
                           X_test = X_valid_comb , 
                          y_train = y_train , 
                          y_test = y_valid)
model_display = pd.DataFrame(model_scores)
model_display
```
